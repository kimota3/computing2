# 다변수 함수


두 개의 독립 변수 $x, y$ 에 대응하여 하나의 변수 $z$ 가 정해지는 함수 $f$ 를

$$
z=f(x, y)
$$

로 나타내자. 마찬가지로 세 개의 독립변수 $x, y, z$ 에 대응하여 하나의 변수 $w$ 가 정해지는 함수 $g$ 를 생각하여

$$
w=g(x, y, z)
$$

로 나타낼 수 있다. 이 장에서는 이와 같이 두 개 이상의 독립 변수를 갖는 함수들을 다룰 것인데 위에 서의 함수 $f$ 와 $g$ 를 정의역에 해당하는 독립변수의 개수에 따라 각각 2 변수함수 그리고 3 변수함수라 한 다. 일반적으로 다음과 같이 정의한다.

**[정의]** (다변수 함수)

일반적으로 $n$ 개의 독립변수 $x_{1}, x_{2}, \cdots, x_{n}$ 에 하나의 변수 $y$ 가 대응하는 함수 $f$ 를

$$
y=f\left(x_{1}, x_{2}, \cdots, x_{n}\right)
$$

로 나타내고 이를 $n$-변수함수(function of $n$ variables)라 한다. 

정의역에 해당하는 두 개 이상의 독립변수를 갖는 함수를 통틀어 다변수함수(function of several variables)라 한다.

**[예제]**


$f(x, y)=x^{2}-2xy+3y^{2}$ 에 대한 다음 값들을 구하여라.

(1) $f(3,0)$

(2) $f(-1,2)$

|풀이|

(1) $f(3,0)=3^{2}-2 \cdot 3 \cdot 0+3 \cdot 0^{2}=9-0+0=9$

(2) $f(-1,2)=(-1)^{2}-2 \cdot(-1) \cdot 2+3 \cdot 2^{2}=1+4+12=17$



**[예제]**

다음 2 변수함수가 정의 되는 영역을 구하여라.

(1) $f(x, y)=\frac{x-y}{x y}$

(2) $f(x)=\sqrt{x-y^{2}}$

|풀이|

(1) $ x y \neq 0$ 이어야 하므로 정의역은 이 $\quad\{(x, y) \mid x \neq 0$ and $y \neq 0\}$

즉 $x$ 축과 $y$ 축을 제외한 $\mathrm{R}^{2}$ 상의 모든 점의 집합이다.

(2) $x-y^{2} \geqq 0$ 이어야 하므로 정의역은
$$
\left\{(x, y) \mid y^{2} \leqq x\right\}
$$

즉 포물선 $x=y^{2}$ 을 포함한 오른쪽 영역이다.

# 이변수 함수의 그래프

이변수 함수 $z=f(x, y)$ 의 그래프는 3 차원 공간 내의 $(x, y, f(x, y))$ 형태의 점들의 집합으로

$$
\left\{(x, y, z) \in \mathbb{R}^{3} \mid z=f(x, y)\right\}
$$
이다. 예제들을 통하여 이변수 함수의 그래프를 그려보자.

**[예제]**

이변수 함수 $f(x,y)= x^2 + y^2$의 그래프를 그려라. 


```python
# [예제] 2변수 함수 그래프

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


def f(x,y):
    z= x**2 + y**2
    #z= -(x+1)*(x)*(x-1)*(x-2)-(y+1)*(y)
    return z

n = 100
x=np.linspace(-10,10,n)
y=np.linspace(-10,10,n)

from mpl_toolkits.mplot3d import Axes3D

xx,yy = np.meshgrid(x,y)

plt.figure(figsize = (7,7))
ax = plt.subplot(111, projection='3d')
surf = ax.plot_surface(xx, yy, f(xx,yy), cmap='viridis')
ax.view_init(10,70) 
#fig.colorbar(surf, shrink=0.6, aspect=8)
plt.show()
```

# 편미분


일변수함수 $y=f(x)$ 의 변화율은 변화의 방향이 왼쪽과 오른쪽의 두 방향만 있지만 2 변수함수 $z=f(x, y)$ 의 변화율은 $x y$ 평면 위의 점 $(x, y)$ 에 서 무한히 많은 방향을 따라 그 변화율이 다를 수가 있다. 따라서 2 변수함수 의 도함수는 일반적으로 각 방향에 따른 도함수로 주어지는데 $x$ 축 방향과 $y$ 축 방향의 도함수가 중요한 역할을 한다. 

이 도함수들은 특히 편도함수 (partial derivative)라고 부른다. 

## $x$ 방향 편도함수

2변수함수 $z=f(x, y)$ 이 주어졌을 때 고정된 한 점 $\left(x_{0}, y_{0}\right)$ 에 대하여 극한

$$
\lim _{h \rightarrow 0} \frac{f\left(x_{0}+h, y_{0}\right)-f\left(x_{0}, y_{0}\right)}{h}
$$

이 존재하는 경우 이 극한을 함수 $f(x, y)$ 의 점 $\left(x_{0}, y_{0}\right)$ 에서의 $x$ 에 관한 편미분 계수라 한다. 
 

위의 편미분 계수들이 정의역의 원소 $(x,y)$에 대하여 존재하는 편미분 계수들을 함수값으로 하는 함수들을 편도함수라 하고 다음과 같이 나타낸다. 

$$
\begin{aligned}
&f_{x}(x, y), f_{x}, \frac{\partial f}{\partial x}, \frac{\partial}{\partial x} f(x, y), \frac{\partial z}{\partial x}, z_{x}, D_{x} f(x, y) \\
&\text { 여기서 기호 } \frac{\partial f}{\partial x} \text { 는 "round } f \text { round } x \text { "라고 읽는다. }
\end{aligned}
$$

## $y$ 방향 편도 함수

$y$ 에 관해서도 비슷하게 극한
$$
\lim _{h \rightarrow 0} \frac{f\left(x_{0}, y_{0}+h\right)-f\left(x_{0}, y_{0}\right)}{\underline{h}}
$$
이 존재하면 이 극한을 함수 $f(x, y)$ 의 $y$ 에 관한 편미분계수라 한다. 그 편미분계수들은 함숫값으로 하는 함수를 $f(x, y)$ 의 $y$ 에 관한 편도함수라 하고 다음 기호 중의 하나로 나타내어 진다. 
$$
f_{y}(x, y), f_{y^{2}}, \frac{\partial f}{\partial y}, \frac{\partial}{\partial y} f(x, y), \frac{\partial z}{\partial y}, z_{y}, D_{y} f(x, y)
$$



**[예제]**

다음 2변수 함수들의 편도함수 $f_{x}(x, y)$ 와 $f_{y}(x, y)$ 를 구하여라.

(1) $f(x, y)=x^{2}+x y+y^{2}$ 

(2) $e^{xy}$



## $n$ 계 편도함수

2변수함수 $z=f(x, y)$ 의 편도함수 $f_{x}(x, y), f_{y}(x, y)$ 는 또 하나의 2 변 수함수이므로 이 함수의 편도함수가 또 존재하면 이것을 함수 $z=f(x, y)$ 의 2계 편도함수(second-order partial derivative)라 한다.

$$
\begin{aligned}
\frac{\partial}{\partial x}\left(\frac{\partial z}{\partial x}\right) &=\frac{\partial^{2} z}{\partial x^{2}},\left(z_{x}\right)_{x}=z_{x x},\left(f_{x}\right)_{x}=f_{x x} \\
\frac{\partial}{\partial y}\left(\frac{\partial z}{\partial x}\right) &=\frac{\partial^{2} z}{\partial y \partial x},\left(z_{x}\right)_{y}=z_{x y},\left(f_{x}\right)_{y}=f_{x y} \\
\frac{\partial}{\partial x}\left(\frac{\partial z}{\partial y}\right) &=\frac{\partial^{2} z}{\partial x \partial y},\left(z_{y}\right)_{x}=z_{y x},\left(f_{y}\right)_{x}=f_{y x} \\
\frac{\partial}{\partial y}\left(\frac{\partial z}{\partial y}\right) &=\frac{\partial^{2} z}{\partial y^{2}},\left(z_{y}\right)_{y}=z_{y y},\left(f_{y}\right)_{y}=f_{y y}
\end{aligned}
$$

마찬가지로 $n$ 계 편도함수 $(n$-th order partial derivative)도 비슷하게 나타 낼 수 있다.

## 편미분 계수의 기하학적 의미

좌표공간에서 2 변수함수 $z=f(x, y)$ 의 그래프와 평면 $y=b$ 가 만나는 부분은 일반적으로
$$
z=f(x, b)
$$
와 같은 곡선의 방정식으로 나타낼 수 있다. 이때, 편미분계수 $f_{x}(a, b)$ 는 곡선 $z=f(x, b)$ 의 $x=a$ 에서의 접선의 기울기이다. 비슷하게 $f_{y}(a, b)$ 는 곡선 $z=f(a, y)$ 의 $y=b$ 에서의 접선의 기울기이다. 

# 2변수 함수의 최댓값과 최솟값 


이 절에서는 편도함수를 이용하여 2 변수함수의 극댓값과 극솟값 그리고 최댓값과 최솟값에 대하여 알아본다.

## 극대/극소

**[정의]**

- 2변수함수 $f(x, y)$ 가 정의역 안의 점 $(a, b)$ 의 어떤 근방의 모든 점 
 $(x, y)$ 에 대해 
 $$f(x, y) \leq f(a, b)$$ 
 
 일 때, $(a, b)$ 를 함수 $f$ 의 극대점이라 하고 이때, $f(a, b)$ 를 $f$ 의 극댓값 (relative maximum)이라 한다. 


- 점 $(a, b)$ 의 어떤 근방의 모든 점 $(x, y)$ 에 대해 
$$f(x, y) \geq f(a, b)$$ 

일 때, $(a, b)$ 를 함수 $f$ 의 극소점이라 하고 이때, $f(a, b)$ 를 $f$ 의 극솟값 (relative minimum)이라 한다. 


- 극대점과 극소점을 통틀어 극점이라고 하고 극댓값과 극솟값을 통틀어 극값이라고 부른다.

**[정리]** 

점 $(a, b)$ 가 2 변수함수 $f(x, y)$ 의 극점이고 각 편도함수 $f_{x}(a, b)$, $f_{y}(a, b)$ 가 존재하면

$$
f_{x}(a, b)=0 \text { 과 } f_{y}(a, b)=0
$$

이 동시에 성립한다.


[참고]

- 위 정리의 역은 반드시 성립하는 것은 아니다. 즉, 편미분 가능한 함수에서 $f_{x}(a, b)=f_{y}(a, b)$ $=0$ 은  함수 $f$ 가 $(a, b)$ 에서 극값을 갖기 위한 필요 조건이다.  


- 따라서 편미분 계수가 존재하지 않는 점들까지 고려하여 극값을 갖을 수 있는 가능성이 있는 점들을 아래와 같은 정의를 통하여 모을 수 있다. 

**[정의]**

함수 $f(x, y)$ 가 점 $(a, b)$ 에서 $f_{x}(a, b)=f_{y}(a, b)=0$ 이거나 두 편도함수 가운데 적어도 하나가 존재하지 않을 때 점 $(a, b)$ 를 함수 $f$ 의 임계점 (critical point)라 한다.

**[예제]**

함수 $f(x, y)=x^{3}-3 x y+y^{3}$ 의 임계점을 모두 구하여라.

|풀이|

함수 $f(x, y)$ 는 모든 점에서 편미분 가능하므로 아래를 해를 구하면 된다. 

$$
f_{z}(x, y)=f_{y}(x, y)=0.
$$

이것으로부터
$$
\begin{aligned}
&x^{2}=y \text { and } x=y^{2} \\
&\Rightarrow x=\left(x^{2}\right)^{2} \\
&\Rightarrow x^{4}-x=0 \Rightarrow x(x-1)\left(x^{2}+x+1\right)=0 \\
&\Rightarrow x=0 \text { or } x=1 \\
&\Rightarrow(x=0 \text { and } y=0) \text { or }(x=1 \text { and } y=1)
\end{aligned}
$$

따라서 임계점은 $(0,0)$ 과 $(1,1)$ 이다.

**[예제]**

(1) 함수 $f(x, y)=y^{2}-x^{2}$ 의 임계점을 구하여라. 

(2) 위에서의 임계점 중 극값이 되는 것은?

|플이|

(1)
$$
\begin{aligned}
&0=f_{x}(x, y)=-2 x \\
&0=f_{y}(x, y)=2 y
\end{aligned}
$$
로부터 임계점 $(0,0)$ 을 얻는다. 

(2) $f(x, 0)=-x^{2}$ 이므로 점 $(0,0)$ 는 극소점이 아니을 알 수 있다.

   $f(0, y)=y^{2}$ 이므로 점 $(0,0)$ 는 극대점 도 될 수 없다. 즉, 극값은 없다.



[참고]
- 위의 예제에서처럼 임계점이지만 극점이 아닌 점을 안장점(saddle point)라고 한다. 
  
- 일변수함수에서 2계 도함수로 함수의 극값을 결정하는 방법이 있었는데 2변수함수에서도 
2계편도함수를 이용하여 극값의 성질을 알아볼 수 있다. 


**[정리]**

2변수함수 $f(x, y)$ 가 점 $(a, b)$ 근방에서 주어져서

$$
f_{z}(a, b)=0, f_{y}(a, b)=0
$$

이 성립하고 2계 도함수들에 대해서 판별식

$$
D=f_{x x}(a, b) f_{yy}(a, b)-\left\{f_{x y}(a, b)\right\}^{2}
$$

이 주어짔을 때,

(1) 만약 $D>0, f_{x x}(a, b)<0$ 이면, 점 $(a, b)$ 는 $f$ 의 극대점이다.

(2) 만약 $D>0, f_{x x}(a, b)>0$ 이면, 점 $(a, b)$ 는 $f$ 의 극소점이다.

(3) 만약 $D<0$ 이면, 점 $(a, b)$ 는 $f$ 의 안장점이다.

(4) 만약 $D=0$ 이면, 이 방볍으로는 판정할 수 없다.





[참고]

- 2변수함수 $f(x, y)$ 가 정의역 안의 모든 점 $(x, y)$ 에 대해 $f(x, y) \leq$ $f(a, b)$ 읻 때, 
 함수 $f(x, y)$ 는 점 $(a, b)$ 에서 최댓값 $f(a, b)$ 을 갖는다고 한다.
 

- 비슷하게 모든 점 $(x, y)$ 에 대해 $f(x, y) \geq f(a, b)$ 읻 때, 점 $(a, b)$ 에서 최솟값 $f(a, b)$ 을 
갖는다고 한다. 


- 일변수함수에서와 마찬가지로 함수 $f(x, y)$ 의 정의역이 유계이고 폐집합이며 $f(x, y)$ 이 연속함수이면 $f(x, y)$ 는 
 그 정의역에서 반드시 최댓값과 최솟값을 갖게 된다.



**[예제]**

함수 $f(x, y)=x^{2}+y^{2}$ 의 극값을 구하여라.

|풀이| 

(1) 극점이 되기 위해서는 최소한 임계점이 되어야 하므로 우섬 임계점을 구하자.

$$
\begin{aligned}
&0=f_{x}(x, y)=2 x \\
&0=f_{y}(x, y)=2 y
\end{aligned}
$$

로부터 임계점 $(0,0)$ 을 얻는다. 

(2)
$$
f_{xx}(0,0)=2, \quad f_{yy}(0,0)=2, \quad f_{xy}(0,0)=0
$$

이므로


$$
D=4>0 \text { 이고 } f_{xx}(0,0)=2>0
$$

이다. 

정리로부터 임계점 $(0,0)$ 은 극소점임을 알 수 있다. 이때, 극솟값은 $f(0,0)=0$ 이다.

연쇄 법칙은 음함수 미분법 과정을 설명하는데 이용된다. $x$ 의 미분가능 함수로써 $y$ 가 $F(x, y)=0$ 형태의 방정식에 의하여 음함수로 정의된다고 가정하자. 다시 말하면, $f$ 의 정의역에 있는 모든 $x$ 에 관하여, $y=f(x)$ 은 $F(x, f(x))=0$ 을 의미한다. 만약 $F$ 가 미분가능하면, 연쇄 법칙을 적용하 여 $x$ 에 관하여 식 $F(x, y)=0$ 의 양변을 미분할 수 있다. $x$ 와 $y$ 의 양 쪽 모두가 $x$ 의 함수이므로 다음을 얻는다.
$$
\frac{\partial F}{\partial x} \frac{d x}{d x}+\frac{\partial F}{\partial y} \frac{d y}{d x}=0
$$
$\frac{d x}{d x}=1$ 이므로, 만약 $\frac{\partial F}{\partial y} \neq 0$ 이면 $\frac{d y}{d x}$ 에 관하여 풀어서 다음을 얻는다.

$$
\frac{d y}{d x}=-\frac{\frac{\partial F}{\partial x}}{\frac{\partial F}{\partial y}}=-\frac{F_{x}}{F_{y}}
$$
$F(x, y)=0$ 이 $y$ 를 $x$ 의 음함수로 정의할 수 있기 위한 조건은 고등미적 분학에서 나오는 음함수정리에서 주어지는데 그 조건은 "F가 $(a, b)$ 를 포함 하는 열린 원판에서 정의되고, $F(a, b)=0, F_{y}(a, b) \neq 0$ 이고, $F_{x}$ 와 $F_{y}$ 가 열린 원판 위에서 연속이다.


예제 $14.15$ $x^{3}+y^{3}=6 x y$ 일 때 $y^{\prime}$ 을 구하여라.

| 풀이 |
주어진 식은 다음과 같이 쓰여질 수 있다.
$$
F(x, y)=x^{3}+y^{3}-6 x y=0
$$
따라서 주어진 공식에 의하여
$$
\frac{d y}{d x}=-\frac{F_{x}}{F_{y}}=-\frac{3 x^{2}-6 y}{3 y^{2}-6 x}=-\frac{x^{2}-2 y}{y^{2}-2 x}
$$
이다.



# 경도 벡터와 등위면

**[정의]** (경도벡터와 등위면)

- 다변수 함수 $f(x, y)$에서 $x$ 편도함수와 $y$ 편도함수들 로 이루어진 벡터 
$$\nabla f=\left(\frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}\right)$$

를  $f$ 의 경도 (=경사진 정도)또는 경도벡터라고 한다. 

- 주어진 $c$를 다변수 함수 $f$의 함수 값으로 갖는 정의역에서의 원소들로 이루어진 집합을 함수 $f$의 $c-$ 등위 집합(level set)이라 한다. 

[참고] 일반적인 $n$ 변수 함수에 대해서도 마찬가지로 정의할 수 있다. 

2변수함수 $f(x, y)$의 경우 등위 집합은 등위선(level curve)이 되며 다음과 같은 사실을 알 수 있다.

**[정리]**

$f(x, y)=c$ 의 접선은 경도와 수직을 이룬다. 

|증명|

- 매개변수 $t$ 를 가진 어떤 함수 벡터 $\mathbf{r}(t)=(g(t), h(t))$ 가 그 등위선 위를 지나는 경우는
$$
f(g(t), h(t))=c
$$
가 만족된다. 

연쇄법칙을 사용하여 위 식의 양변을 $t$ 에 관하여 미분하면

$$
\frac{\partial f}{\partial x} g^{\prime}(t)+\frac{\partial f}{\partial y} h^{\prime}(t)=0
$$

이 되고 따라서

$$
\begin{gathered}
\frac{\partial f}{\partial x} g^{\prime}(t)+\frac{\partial f}{\partial y} h^{\prime}(t)=0 \\
\left(\frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}\right) \cdot\left(g^{\prime}(t), h^{\prime}(t)\right)=0 \\
\nabla f \cdot \mathbf{r}^{\prime}(t)=0
\end{gathered}
$$

이 된다. 

- 그러므로 경도벡터 $\nabla f(x, y)$ 는 등위선의 접선멕터 $\mathbf{r}^{\prime}(t)$ 와 수직 이다.

[참고] 마찬가지로 3변수 함수에서는 등위집합이 등위면이 되면 경도벡터는 접평면과 수직임을 증명할 수 있다. 



```python
import numpy as np
import matplotlib.pyplot as plt

def j(x,y):
    z=x*y    
    return z

x_2=np.linspace(-2,2,21)
y_2=np.linspace(-2,2,21)

X_2,Y_2 = np.meshgrid(x_2,y_2)

#print(X_2)
#print(Y_2)

Z_2 = j(X_2,Y_2)

CC = plt.contour(X_2,Y_2,Z_2,levels=[-1,-0.5,0,0.5,1],colors='black')

plt.clabel(CC,inline=2,fontsize = 9)

#plt.colorbar
plt.xlabel("$x$")
plt.ylabel("$Y$")
plt.show()
```

# 라그랑주 승수법


## 등식 제한 조건을 갖는 이변수 함수의 최대/최소


[목적]  $g(x, y)=k$ 와 같은 형태의 제약조건이 있는 경우 이변수 함수 $f(x, y)$의 최댓값 또는 최솟값을 구하는 라그랑주의 방법을 소개한다. 

- 그림에서 곡선 $g(x, y)=k$ 를 $f$ 의 여러 개의 등위곡선들과 함께 보여준다. 
이들은 $c=7,8,9,10,11$ 에서의 방정식 $f(x, y)=c$ 를 나타낸다. 


- $g(x, y)=k$ 형태의 조건을 만족하는 $f(x, y)$의 최댓값은 $g(x, y)=k$ 와 교차하는 등위곡선 $f(x, y)=c$ 에서 $c$ 의 최댓값을 구하는 것이다. 그림으로부터 이러한 상황은 곡선이 서로 접할 때 읻어난다는 것을 알 수 있다. 즉, 하나의 공통접선을 가진다. 


- 이것은 그들이 서로 접하는 점 $\left(x_{0}, y_{0}\right)$ 에서 같은 법선을 가짐을 의미한다. 그러므로 기울기 벡터들은 평행하다. 


- 즉, 적당한 스칼라 $\lambda$ 에 대하여 다음 조건이 만족된다.

$$
\nabla f\left(x_{0}, y_{0}\right)=\lambda \nabla g\left(x_{0}, y_{0}\right)
$$


- 위 식에서 $\lambda$ 를 라그랑주 승수(Lagrange Multiplier)라 부른다.




```python
import numpy as np
import matplotlib.pyplot as plt

# 목적함수 f(x) = x1^2 + x2^2
def f1(x1, x2):
    return x1**2 + x2** 2

x1 = np.linspace(-3.5, 3.5, 100)
x2 = np.linspace(-3.5, 3.5, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = f1(X1, X2)

# 등식 제한조건 방정식 g(x) = x1 + x2 - 1 = 0





plt.figure(figsize=(6, 6))
CC=plt.contour(X1, X2, Y, colors="gray",levels=[0.2,1.0, 2, 4,6])
#plt.plot(x1, x2_g, 'g-')


plt.clabel(CC,inline=2,fontsize = 11)


def f2(x1, x2):
    return x1*x2

Y2 = f2(X1,X2)

plt.contour(X1, X2, Y2, colors='blue',levels=[1])


#plt.plot([0], [0], 'bo')
plt.plot([-1,1], [-1,1], 'ro', ms=5)

#plt.xlim(-5, 5)
#plt.ylim(-5, 5)
#plt.xticks(np.linspace(-3, 4, 9))
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
#plt.title("test")
plt.show()
```


2변수함수에 대한 라그랑주 승수법을 생각애 보자. 제약조건 $g(x, y)=k$ 를 만족하는 $f(x, y)$ 의 최대, 최솟값을 구하려면

$$
\begin{aligned}
\nabla f(x, y) &=\lambda \nabla g(x, y) \\
g(x, y) &=k
\end{aligned}
$$

을 만족하는 $x, y, \lambda$ 를 구해야 한다. 이것은 다음과 같은 세 미지수를 갖는 세 방정식을 연립으로 푸는 읻에 해당된다.

$$
\begin{aligned}
&f_{x}=\lambda g_{z} \\
&f_{v}=\lambda g_{v} \\
&g(x, y)=k
\end{aligned}
$$

- 정확한 $\lambda$ 값을 구하는 것은 필요하지 않을 수 있다. 

**[예제]** 

원 $x^{2}+y^{2}=1$ 위에서 정의된 함수 $f(x, y)=x^{2}+y$ 의 최대, 최솟값을 구하여라.

|풀이|

- 제약조건 $g(x, y)=x^{2}+y^{2}=1$ 를 만족하는 $f$ 의 최대, 최솟값에 대한 문제이다. 

- 라그랑주 송수의 방볍을 이용하여 방정식 $\nabla f=\lambda \nabla g, g(x, y)=1$ 을 푼다. 

$$
\begin{aligned}
&f_{x}=\lambda g_{z} \\
&f_{v}=\lambda g_{v} \\
&g(x, y)=1
\end{aligned}
$$

즉,

$$
\begin{aligned}
&2 x=2 x \lambda \\
&1=2 y \lambda \\
&x^{2}+y^{2}=1
\end{aligned}
$$

이다. 

- $2 x=2 x \lambda$ 로부터 $(x=0$ 또는 $\lambda=1$ ) 을 얻는다. 만약 $x=0$ 이면 $x^{2}+y^{2}=1$ 에서 $y=\pm 1$ 이다.

- 만약 $\lambda=1$ 이면 $1=2 y \lambda$ 로부터 $y=1 / 2$ 이고 따라서 $x^{2}+y^{2}=1$ 로부터 $x=\pm \sqrt{3} / 2$ 이다.
그러므로 $f$ 는 점 $(0,1),(0,-1),(\sqrt{3} / 2,1 / 2),(-\sqrt{3} / 2,1 / 2)$ 에서 최댓값, 최솟값을 가질 수 있다. 

- 위 네 점에서 $f$ 의 값을 구하면 다음을 얻는다.

$$
\begin{aligned}
&f(0,1)=1 \\
&f(0,-1)=-1 \\
&f(\pm \sqrt{3} / 2,1 / 2)=5 / 4
\end{aligned}
$$

그러므로 $x^{2}+y^{2}=1$ 위에서 $f$ 의 최댓값은 $f(\pm \sqrt{3} / 2,1 / 2)=5 / 4$ 이고, 최솟값은 $f(0,-1)=-1$ 이다. 

```python
import numpy as np
import matplotlib.pyplot as plt

# 목적함수 f(x) = x1^2 + x2^2
def f1(x1, x2):
    return x1 ** 2 + x2
x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = f1(X1, X2)

def f2(x, y):
    return x ** 2 + y**2
Y2 = f2(X1,X2)
# 등식 제한조건 방정식   x1**2 + x2**2 = 1
#x2_g = np.sqrt(+1 - x1**2)
#x3_g = -np.sqrt(+1 - x1**2)



plt.figure(figsize=(6, 6))
CC=plt.contour(X1, X2, Y, colors="gray", levels=[-1,0.5, 1, round(5/4,2),2,4])
plt.clabel(CC,inline=10,fontsize = 11)

#plt.colorbar

plt.contour(X1, X2, Y2,colors="blue", levels=[1])

#plt.plot([0,0], [-1,1], 'ro', ms=5)

#plt.plot([], [], 'rP')
#plt.plot([], [], 'ro', ms=10)

#plt.xlim(-3, 1)
#plt.ylim(-1, 1)
#plt.xticks(np.linspace(-3, 3, 9))
plt.xlabel("$x$")
plt.ylabel("$y$")
#plt.title(" ")
plt.show()
```

## 등식 제한 조건을 갖는 3변수 함수의 최대/최소

[목적]  $g(x, y, z)=k$ 와 같은 형태의 제약조건이 있는 경우 3변수 함수 $f(x, y, z)$의 최댓값 또는 최솟값을 조사한다.

- 2변수함수에서의 논법이 그대로 적용된다. 


- 그림에서의 등위곡선들 대신 등위곡면 $f(x, y, z)=c$ 를 생각한다. 


- $f$ 의 최댓값이 $f\left(x_{0}, y_{0}, z_{0}\right)=c$ 이면 등위곡 면 $f(x, y, z)=c$ 는 등위곡면 $g(x, y, z)=k$ 에 접하게 되고, 따라서 대응되는 기울기 벡터들은 서로 평행하다. 


- 그러므로 만약 $\nabla g\left(x_{0}, y_{0}, z_{0}\right) \neq 0$ 이면 다음 조건을 만족하는 실수 $\lambda$ 가 존재한다.


$$
\nabla f\left(x_{0}, y_{0}, z_{0}\right)=\lambda \nabla g\left(x_{0}, y_{0}, z_{0}\right)
$$


제약조건 $g(x, y, z)=k$ 를 만족하는 $f(x, y, z)$ 의 최댓간과 최솟값을 구하 기 위해 $($ 이들 극값이 존재하고 곡면 $g(x, y, z)=k$ 위에서 $\nabla g \neq 0$ 한다고 가정함)

- 다음을 만족하는 모든 $x, y, z$ 와 $\lambda$ 의 값을 구한다.

$$
\begin{aligned}
\nabla f(x, y, z) &=\lambda \nabla g(x, y, z) \\
g(x, y, z) &=k
\end{aligned}
$$

- 위에서 구한 모든 점 $(x, y, z)$ 에서 $f$ 의 값을 계산한다. 이 값들 중 가장 큰 것이 $f$ 의 최댓값이고, 가장 작은 것이 $f$ 의 최솟값이다.


만약, 벡터식 $\nabla f=\lambda \nabla g$ 를 그들의 성분으로 쓰면 다음과 같다.

$$
\begin{aligned}
&f_{x}=\lambda g_{x} \\
&f_{y}=\lambda g_{v} \\
&f_{z}=\lambda g_{z} \\
&g(x, y, z)=k
\end{aligned}
$$

**[예제]** 

뚜겅이 없는 직욕면체의 상자가 $12 \mathrm{~m}^{2}$ 의 넓이의 판지로 만들어졌다. 이 상 자의 부피의 최댓값을 구하여라.

|풀이|

$x, y, z$ 를 각가 길이, 폭, 뉴이라 하자. 제약조건

$$
g(x, y, z)=2 x z+2 y z+x y=12
$$

하에서 $V=x y z$ 를 최대화하려고 한다. 라그랑주 승수법을 이용하여

$$
\begin{aligned}
&\nabla V=\lambda \nabla g \\
&g(x, y, z)=12
\end{aligned}
$$

을 만족하는 $x, y, z, \lambda$ 를 구하여 보자. 즉,

$$
\begin{aligned}
&V_{x}=\lambda g_{z} \\
&V_{y}=\lambda g_{y} \\
&V_{x}=\lambda g_{z} \\
&2 x z+2 y z+x y=12
\end{aligned}
$$

이들은 다음과 같이 된다.
$$
\begin{aligned}
&y z=\lambda(2 z+y) \\
&x z=\lambda(2 z+x) \\
&x y=\lambda(2 x+2 y) \\
&2 x z+2 y z+x y=12
\end{aligned}
$$

여기에서 $\lambda \neq 0$ 이다. 왜냐하면 $\lambda=0$ 이면 $y z=x z=x y=0$ 이 되고 $2 \times 0+2 \times 0+0=12$ 으로 모순이 되기 때문이다. 위 연립방정식의 처음 세 식에 $x, y, z$ 를 가각 곱하면 다음을 얻는다.

$$
\begin{aligned}
&x y z=\lambda(2 x z+x y) \\
&x y z=\lambda(2 y z+x y) \\
&x y z=\lambda(2 x z+2 y z)
\end{aligned}
$$

따라서 처음 두 식으로부터

$$
\lambda(2 x z+x y)=\lambda(2 y z+x y)
$$

이다. 여기에서 $\lambda \neq 0$ 이므로 $x z=y z$ 를 얻는다. 그더나 $z \neq 0(z=0$ 이면 $V=0$ 이다)이므로 $x=y$ 이다. 

- 마찬가지로 $x \neq 0$ 이고
$$
\lambda(2 y z+x y)=\lambda(2 x z+2 y z)
$$

로부터 $y=2 z$ 를 얻는다. 이제 $2 x z+2 y z+x y=12$ 에 $x=y=2 z$ 를 대 입하면

$$
4 z^{2}+4 z^{2}+4 z^{2}=12
$$

를 얻는다. $x, y, z$ 모두 양수이므로 $z=1, x=2, y=2$ 를 얻는다.

  음함수 미분법
